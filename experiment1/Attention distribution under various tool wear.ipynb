```python
#注意力权重显示
from matplotlib.collections import LineCollection
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import normalize
from matplotlib.patches import Rectangle


HIGHFREQ_INPUT_SIZE = [2304, 6]
LOWFREQ_INPUT_SIZE = [288, 7]
ATTENTION_RANGE = 48
PADDING_LENGTH = int(ATTENTION_RANGE - HIGHFREQ_INPUT_SIZE[0]/LOWFREQ_INPUT_SIZE[0])
sample_list = np.arange(428,830,15)

TESTDATA_LOADPATH = 'tool_wear_data_5_small/test_data.npy' #验证集数据1读取路径
TESTLABEL_LOADPATH =  'tool_wear_data_5_small/test_label.npy'

ATT_LOWHIGH_LOADPATH = 'attention_distribution_dataset8/attention_lowhigh_1.npy'
ATT_DETAIL_LOADPATH = 'attention_distribution_dataset8/attention_detail_1.npy'

ATT_DIFF_SAVEPATH = 'attention_distribution_dataset8/attention_translation_differnece.png'
SIG_SAVEPATH = 'attention_distribution_dataset8/translated_signal.png'

Y_test = np.load(TESTLABEL_LOADPATH)
plt.plot(Y_test[428:855])
plt.show()
print(np.shape(Y_test))

for sample_index in sample_list:
    print(sample_index)
    X_test = np.load(TESTDATA_LOADPATH)[sample_index]
    att_lowhigh = np.load(ATT_LOWHIGH_LOADPATH)[sample_index]
    att_detail = np.load(ATT_DETAIL_LOADPATH)[sample_index]

    fig_x = 19
    fig_y = 3.5
    plt.figure(figsize=(fig_x, fig_y))
    plt.subplots_adjust(wspace=0, hspace=0)
    plt.xticks([])
    plt.yticks([])
    plt.xlim(0, np.shape(X_test)[0])
    plt.subplot(2, 1, 1)
    plt.plot(X_test[:, 0], lw=1, color='black')
    plt.subplot(2, 1, 2)
    plt.plot(X_test[:, 3], lw=1, color='black')
    plt.savefig(SIG_SAVEPATH, dpi = 600, bbox_inches='tight')
    plt.show()
    
    fig_x = 25
    fig_y = 3
    plt.figure(figsize=(fig_x, fig_y))
    plt.subplots_adjust(wspace=0, hspace=0)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(att_detail.reshape([8,288]), cmap='Greys')
    plt.savefig(ATT_DIFF_SAVEPATH, dpi = 600, bbox_inches='tight')
    plt.show()


    fig_x = 20
    fig_y = 13
    plt.figure(figsize=(fig_x, fig_y))
    plt.subplots_adjust(wspace=0, hspace=0)

    plt.subplot(2, 3, 1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(att_detail.reshape([48,48]), cmap='Greys')

    plt.savefig(ATT_DIFF_SAVEPATH, dpi = 600, bbox_inches='tight')
    plt.show()
```
