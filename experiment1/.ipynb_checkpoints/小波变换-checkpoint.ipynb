```python
import pywt
import numpy as np
from matplotlib import pyplot as plt
from pywt._doc_utils import boundary_mode_subplot

#小波 https://pywavelets.readthedocs.io/en/latest/ref/wavelets.html
print(pywt.families()) #输出内置小波族
print(pywt.wavelist(kind='discrete'))
# print(pywt.wavelist('morl')) #输出小波列表
wavelet = pywt.Wavelet('db1') #定义小波对象，包含各种属性
print(wavelet)
phi, psi, x = wavelet.wavefun(level=5) #计算给定细化级别的父函数phi，母函数psi的近似值，相应的x坐标

#信号的扩展模式 https://pywavelets.readthedocs.io/en/latest/ref/signal-extension-modes.html

print('pywt.Modes.modes', pywt.Modes.modes) #信号扩展模式
#各种扩展模式的直观展示
signal = 5 - np.linspace(-1.9, 1.1, 9)**2   #信号
fig, axes = plt.subplots(1, 3, figsize=(15, 3))
plt.subplots_adjust(hspace=0.5)
axes = axes.ravel()
boundary_mode_subplot(signal, 'symmetric', axes[0], symw=False)
plt.show()
```

```python
# 一维离散小波变换

import pywt
import numpy as np
from matplotlib import pyplot as plt
from pywt._doc_utils import boundary_mode_subplot

signal = 5 - np.linspace(-1.9, 1.1, 9)**2   #信号
print('signal', signal)

#离散小波变换 https://pywavelets.readthedocs.io/en/latest/ref/dwt-discrete-wavelet-transform.html
# 单级
coeffs1 = pywt.dwt(signal, 'db1', 'smooth')
print('单级离散小波变换', coeffs1)
# 多级
coeffs2 = pywt.wavedec(signal, 'db1', 'smooth', level=2)
print('多级离散小波变换', coeffs2)
# 有意义的最大分解级数
print(pywt.dwtn_max_level((64, 32), 'db2')) #高维

# 离散小波逆变换 https://pywavelets.readthedocs.io/en/latest/ref/idwt-inverse-discrete-wavelet-transform.html
signal_rec = pywt.idwt(coeffs1[0], coeffs1[1], 'db1', 'smooth') #单级
print('signal_rec', signal_rec)
signal_rec = pywt.waverec(coeffs2, 'db1', 'smooth') #多级
print('signal_rec', signal_rec)
signal_rec = pywt.upcoef('a', coeffs1[0], 'db1') + pywt.upcoef('d', coeffs1[1], 'db1') #部分
print('signal_rec', signal_rec)
```

```python
#二维离散小波变换 https://pywavelets.readthedocs.io/en/latest/ref/2d-dwt-and-idwt.html

import numpy as np
import pywt
from matplotlib import pyplot as plt

image = pywt.data.camera().astype(np.float32) #信号

coeffs1, coeffs2 = pywt.dwt2(image, 'haar', mode='symmetric') #单级小波变换，输出1低频和3个高频

image_rec = pywt.idwt2([coeffs1, coeffs2], 'haar', mode='symmetric') #单级小波逆变换
plt.imshow(image_rec, cmap='gray')
plt.show()

coeffs = pywt.wavedec2(image, 'db1', mode='symmetric', level=3) #多级小波变换
print(np.shape(coeffs))
pywt.waverec2(coeffs, 'db1', mode='symmetric') #多级小波逆变换

# 离散小波系数处理 https://pywavelets.readthedocs.io/en/latest/ref/dwt-coefficient-handling.html
arr, coeff_slices = pywt.coeffs_to_array(coeffs) #n维小波系数拼接
print(np.shape(arr))
plt.imshow(arr, cmap='gray')
plt.show()
```

```python
#二维小波分解示意图 https://pywavelets.readthedocs.io/en/latest/ref/2d-decompositions-overview.html

import numpy as np
import pywt
from matplotlib import pyplot as plt
from pywt._doc_utils import wavedec2_keys, draw_2d_wp_basis

label_levels = 3  # how many levels to explicitly label on the plots

x = pywt.data.camera().astype(np.float32)

fig, axes = plt.subplots(2, 4, figsize=[14, 8])
for level in range(0, label_levels + 1):
    if level == 0:
        axes[0, 0].set_axis_off()
        axes[1, 0].imshow(x, cmap=plt.cm.gray)
        axes[1, 0].set_title('Image')
        axes[1, 0].set_axis_off()
        continue

    # plot subband boundaries of a standard DWT basis
    draw_2d_wp_basis(x.shape, wavedec2_keys(level), ax=axes[0, level],
                     label_levels=label_levels)
    axes[0, level].set_title('{} level\ndecomposition'.format(level))

    # compute the 2D DWT
    c = pywt.wavedec2(x, 'db2', mode='periodization', level=level)
    # normalize each coefficient array independently for better visibility
    c[0] /= np.abs(c[0]).max()
    for detail_level in range(level):
        c[detail_level + 1] = [d/np.abs(d).max() for d in c[detail_level + 1]]
    # show the normalized coefficients
    arr, slices = pywt.coeffs_to_array(c)
    axes[1, level].imshow(arr, cmap=plt.cm.gray)
    axes[1, level].set_title('Coefficients\n({} level)'.format(level))
    axes[1, level].set_axis_off()

plt.tight_layout()
plt.show()
```

```python
#一维连续小波变换 https://pywavelets.readthedocs.io/en/latest/ref/cwt.html

import pywt
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(256)
y1 = np.sin(2*np.pi*x/32)
y2 = np.sin(2*np.pi*x/128)
y = np.concatenate([y1, y2], axis=0)
x = np.arange(512)
print(np.shape(y))

plt.plot(x, y)
coef, freqs=pywt.cwt(y,np.arange(1,129),'gaus1')
plt.matshow(coef) # doctest: +SKIP
plt.show() # doctest: +SKIP

t = np.linspace(-1, 1, 200, endpoint=False)
sig  = np.cos(2 * np.pi * 7 * t) + np.real(np.exp(-7*(t-0.4)**2)*np.exp(1j*2*np.pi*2*(t-0.4)))
plt.plot(sig)
plt.show()
widths = np.arange(1, 31)
cwtmatr, freqs = pywt.cwt(sig, widths, 'mexh')
plt.imshow(cwtmatr, extent=[-1, 1, 1, 31], cmap='PRGn', aspect='auto',
           vmax=abs(cwtmatr).max(), vmin=-abs(cwtmatr).max())  # doctest: +SKIP
plt.show() # doctest: +SKIP
```

```python
# 小波包变换 https://blog.csdn.net/qq_40587575/article/details/83154042
import pywt
import numpy as np
import matplotlib.pyplot as plt

level = 5

x = np.arange(256)
y1 = np.sin(2*np.pi*x/4)
y2 = np.sin(2*np.pi*x/16)
y = np.concatenate([y1, y2], axis=0)
x = np.arange(512)
print(np.shape(y))
plt.figure(figsize=(20,2))
plt.xlim(0, 512)
plt.plot(y)
plt.show()


wp = pywt.WaveletPacket(data=y, wavelet='db5', mode='symmetric', maxlevel=level)

a = ['aaaaa', 'aaaad', 'aaada', 'aaadd', 'aadaa', 'aadad', 'aadda', 'aaddd',
    'adaaa', 'adaad', 'adada', 'adadd', 'addaa', 'addad', 'addda', 'adddd',
    'daaaa', 'daaad', 'daada', 'daadd', 'dadaa', 'dadad', 'dadda', 'daddd',
    'ddaaa', 'ddaad', 'ddada', 'ddadd', 'dddaa', 'dddad', 'dddda', 'ddddd']
input_data = np.zeros([np.int(2 ** level), len(wp[a[0]].data)])
# print(np.shape(input_data))
for i in range(np.int(2 ** level)):
    input_data[i] = wp[a[i]].data
plt.figure(figsize=(20,3))
plt.imshow(abs(input_data), cmap='gray')
```

```python
# 小波包变换 https://blog.csdn.net/qq_40587575/article/details/83154042
import pywt
import numpy as np
import matplotlib.pyplot as plt

level = 6

x = np.arange(0, 0.5, 1/512)
y1 = np.sin(2*np.pi*32*x)
y2 = np.sin(2*np.pi*64*x)
y = np.concatenate([y1, y2], axis=0)
x = np.arange(0, 1, 1/512)
plt.figure(figsize=(20,2))
plt.xlim(0, 1)
plt.plot(x, y)
plt.show()


wp = pywt.WaveletPacket(data=y, wavelet='db3', mode='symmetric', maxlevel=level)
wp_level = wp.get_level(level, order="freq")
print(np.shape(wp_level))
input_data = np.zeros([np.int(2 ** level), len(wp_level[0].data)])
for i in range(np.int(2 ** level)):
    input_data[i] = wp_level[i].data
print(np.shape(input_data))
plt.figure(figsize=(20,3))
plt.plot(abs(input_data), cmap='gray')
```
