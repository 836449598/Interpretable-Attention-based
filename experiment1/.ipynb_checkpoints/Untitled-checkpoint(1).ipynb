```python
# 一维离散小波变换
#可以考虑对不同传感器信号采用不同的频带

import pywt
import numpy as np
import matplotlib.pyplot as plt
from pywt._doc_utils import boundary_mode_subplot
import tensorflow as tf
from tensorflow import keras
import tensorflow.contrib.rnn as rnn
from tensorflow.keras import models, layers, utils, backend
import xlrd
import time
import os

TRAINDATA_LOADPATH = 'tool_wear_data_1/train_data.npy' #训练集数据读取路径
TRAINLABEL_LOADPATH = 'tool_wear_data_1/train_label.npy' #训练集标签读取路径
VALIDATIONDATA_LOADPATH = 'tool_wear_data_1/validation_data.npy' #验证集数据读取路径
VALIDATIONLABEL_LOADPATH =  'tool_wear_data_1/validation_label.npy' #验证集标签读取路径
TESTDATA_LOADPATH = 'tool_wear_data_1/test_data.npy' 
#验证集数据读取路径
TESTLABEL_LOADPATH =  'tool_wear_data_1/test_label.npy' #验证集标签读取路径
SUMMARY_PATH = './logs'     #记录路径

os.environ["CUDA_VISIBLE_DEVICES"] = "-1"# 这一行注释掉就是使用gpu

INPUT_NUM = 7
SCALE_NUM = 5 #小波分解级数

SIZE_RNN1_LOWFREQ = 15
SIZE_RNN2_LOWFREQ = 20
SIZE_LOWFREQ =20
SIZE_RNN1_MIDFREQ = 15
SIZE_RNN2_MIDFREQ = 20
SIZE_MIDFREQ = 20
SIZE_RNN1_HIGHFREQ = 15
SIZE_RNN2_HIGHFREQ = 20
SIZE_HIGHFREQ = 20

SIZE_RNN1_ENTIREFREQ = 25
SIZE_ATTENTION_ENTIREFREQ = 25
SIZE_RNN2_ENTIREFREQ = 30

SIZE_DENSE = 200
SIZE_OUTPUT = 3

BATCH_SIZE = 64
PREDICTOR_TRAIN_BATCH =  10

sample_index = 3000


##############################数据集读取########################################
X_train=np.load(TRAINDATA_LOADPATH)
Y_train=np.load(TRAINLABEL_LOADPATH)
# X_validation=np.load(VALIDATIONDATA_LOADPATH)
# Y_validation=np.load(VALIDATIONLABEL_LOADPATH)
X_test=np.load(TESTDATA_LOADPATH)
Y_test=np.load(TESTLABEL_LOADPATH)

#validation_split将样本集按先后比例分为训练集合样本集，样本数量应为两者和的整数倍
X_train = X_train[:(np.shape(X_train)[0]-np.shape(X_train)[0]%np.int(BATCH_SIZE*1.25)), :,:INPUT_NUM]
X_train = X_train[:28160]
Y_train = Y_train[:(np.shape(Y_train)[0]-np.shape(Y_train)[0]%np.int(BATCH_SIZE*1.25)), :SIZE_OUTPUT]
Y_train = Y_train[:28160]

X_train1 = X_train[:np.int(np.shape(Y_train)[0]/3)]
X_train2 = X_train[np.int(np.shape(Y_train)[0]/3):np.int(np.shape(Y_train)[0]/3*2)]
X_train3 = X_train[np.int(np.shape(Y_train)[0]/3*2):]
print('X_train1', np.shape(X_train1))
print('X_train2', np.shape(X_train2))
print('X_train3', np.shape(X_train3))
print('Y_train', np.shape(Y_train))

# X_validation = X_validation[:(np.shape(X_validation)[0]-np.shape(X_validation)[0]%BATCH_SIZE), :,:INPUT_NUM]
# Y_validation = Y_validation[:(np.shape(Y_validation)[0]-np.shape(Y_validation)[0]%BATCH_SIZE), :SIZE_OUTPUT]
# print('X_validation', np.shape(X_validation))
# print('Y_validation', np.shape(Y_validation))

X_test = X_test[:(np.shape(X_test)[0]-np.shape(X_test)[0]%BATCH_SIZE),:, :INPUT_NUM]
Y_test = Y_test[:(np.shape(Y_test)[0]-np.shape(Y_test)[0]%BATCH_SIZE), :SIZE_OUTPUT]
print('X_test', np.shape(X_test))
print('Y_test', np.shape(Y_test))

###############################################小波分析###############################################################
def normalization(data): #归一化
#     print('data', np.shape(data))
    x = data.reshape([-1, np.shape(data)[-1]])
    max_val = np.max(x, 0)
#     print('max_val', np.shape(max_val))
    min_val = np.min(x, 0)
#     print('min_val', min_val)
    x = (x - min_val)/(max_val - min_val + 1e-8)*1.8 - 0.9
    x = x.reshape(np.shape(data))
#     print('x', np.shape(x))
    return x
def percentile_compute(signal): #通过百分数滤除小尺度部分的噪声，输入小波信号整体
    for i in range(INPUT_NUM):
        signal1 = signal[-1][:,:,i]
        percentile = np.percentile(np.abs(signal1), 90)
        signal1[np.abs(signal1) < percentile] = 0
        signal[-1][:,:,i] = signal1
    for i in range(INPUT_NUM):
        signal1 = signal[-2][:,:,i]
        percentile = np.percentile(np.abs(signal1), 80)
        signal1[np.abs(signal1) < percentile] = 0
        signal[-2][:,:,i] = signal1
#     print('aaa', np.sum(np.abs(signal1) > 0)/np.shape(signal1)[0] / np.shape(signal1)[1]) 
    return signal
def mean_filter(data, ratio):#均值滤波降采样，输入数据和整数压缩比
    index1 = np.int32(np.arange(np.shape(data)[1]* ratio)/ratio)
    if ratio == 0.5:
        return np.mean([np.abs(data[:,index1]), np.abs(data[:,index1+1])], axis = 0)
    if ratio == 0.25:
        return np.mean([np.abs(data[:,index1]), np.abs(data[:,index1+1]), np.abs(data[:,index1+2]), np.abs(data[:,index1+3])], axis = 0)
    else:
        print('error')
def max_filter(data, ratio):#绝对值最大滤波降采样，输入数据和整数压缩比，输出绝对值，如果能输出原值更好
    index1 = np.int32(np.arange(np.shape(data)[1]* ratio)/ratio)
    if ratio == 0.5:
        return np.max([np.abs(data[:,index1]), np.abs(data[:,index1+1])], axis = 0)
    if ratio == 0.25:
        return np.max([np.abs(data[:,index1]), np.abs(data[:,index1+1]), np.abs(data[:,index1+2]), np.abs(data[:,index1+3])], axis = 0)
    else:
        print('error')
def lowfreq_coeffs_construct(coeffs0, coeffs1, coeffs2):#将不同尺度的小波系数降采样为同一长度
    coeffs2 = max_filter(coeffs2, 0.5)
    if (np.shape(coeffs0)[-1] != np.shape(coeffs1)[-1]) or (np.shape(coeffs1)[-1] != np.shape(coeffs2)[-1]):
        print('Length mismatch')
    coeffs = np.concatenate([coeffs0, coeffs1, coeffs2], axis = 2)
    return coeffs
def highfreq_coeffs_construct(coeffs0, coeffs1, coeffs2):#将不同尺度的小波系数降采样为同一长度
    coeffs1 = max_filter(coeffs1, 0.5)
    coeffs2 = max_filter(coeffs2, 0.25)
    if (np.shape(coeffs0)[-1] != np.shape(coeffs1)[-1]) or (np.shape(coeffs1)[-1] != np.shape(coeffs2)[-1]):
        print('Length mismatch')
    coeffs = np.concatenate([coeffs0, coeffs1, coeffs2], axis = 2)
    return coeffs

def wavelet_transform(data):
    coeffs = np.arange(SCALE_NUM+1).tolist()
    for j in range(INPUT_NUM):
        sample = data[:, :, j].reshape([-1, np.shape(data)[1]])
        coeffs2 = pywt.wavedec(sample, 'db8', 'symmetric', level=SCALE_NUM)
        if j == 0:
            for i in range(SCALE_NUM+1):
                coeffs[i] = np.array(coeffs2[i]).reshape([np.shape(sample)[0], 1, np.shape(coeffs2[i])[1]])
        else:
            for i in range(SCALE_NUM+1):
                coeffs[i] = np.concatenate([coeffs[i], np.array(coeffs2[i]).reshape([np.shape(sample)[0], 1, np.shape(coeffs2[i])[1]])], axis = 1)
    coeffs[0] = np.swapaxes(coeffs[0], 1, 2)[:, 6:-8]
    coeffs[1] = np.swapaxes(coeffs[1], 1, 2)[:, 6:-8]
    coeffs[2] = np.swapaxes(coeffs[2], 1, 2)[:, 6:-8]
    coeffs[3] = np.swapaxes(coeffs[3], 1, 2)[:, 6:-7]
    coeffs[4] = np.swapaxes(coeffs[4], 1, 2)[:, 5:-6]
    coeffs[5] = np.swapaxes(coeffs[5], 1, 2)[:, 3:-4]

#     for j in range(1): #系数显示
#         for i in range(1, SCALE_NUM):
#             x1 = np.linspace(0, np.shape(coeffs[i])[1]-1, np.shape(coeffs[i])[1])
#             x2 = np.linspace(0, np.shape(coeffs[i])[1]-1, np.shape(coeffs[SCALE_NUM])[1])
#             if i == 1:
#                 label_interp = np.abs([np.interp(x2, x1, coeffs[i][sample_index, :, j])])
#             else:
#                 label_interp = np.concatenate([label_interp, np.abs([np.interp(x2, x1, coeffs[i][sample_index, :, j])])], axis=0)
#     #     print('label_interp', np.shape(label_interp))
#         plt.figure(figsize=(30,2))
#         plt.contourf(abs(label_interp), cmap='gray')
#         plt.show()
    ######################################################################
    for i in range(6):
        coeffs[i] = normalization(coeffs[i])
        print(i, np.shape(coeffs[i]))
    coeffs = percentile_compute(coeffs)
    
    for i in range(1, SCALE_NUM):  #系数显示
        x1 = np.linspace(0, np.shape(coeffs[i])[1]-1, np.shape(coeffs[i])[1])
        x2 = np.linspace(0, np.shape(coeffs[i])[1]-1, np.shape(coeffs[SCALE_NUM])[1])
        if i == 1:
            label_interp = np.abs([np.interp(x2, x1, coeffs[i][sample_index, :, 0])])
        else:
            label_interp = np.concatenate([label_interp, np.abs([np.interp(x2, x1, coeffs[i][sample_index, :, 0])])], axis=0)
#     print('label_interp', np.shape(label_interp))
    plt.figure(figsize=(30,2))
    plt.contourf(np.abs(label_interp), cmap='gray')
    plt.show()
    
    coeffs_lowfreq = lowfreq_coeffs_construct(coeffs[0], coeffs[1], coeffs[2])
    coeffs_highfreq = highfreq_coeffs_construct(coeffs[3], coeffs[4], coeffs[5])
    return coeffs_lowfreq, coeffs_highfreq

coeffs_lowfreq_train1, coeffs_highfreq_train1 = wavelet_transform(X_train1)
coeffs_lowfreq_train2, coeffs_highfreq_train2 = wavelet_transform(X_train2)
coeffs_lowfreq_train3, coeffs_highfreq_train3 = wavelet_transform(X_train3)
coeffs_lowfreq_train = np.concatenate([coeffs_lowfreq_train1, coeffs_lowfreq_train2, coeffs_lowfreq_train3], axis = 0)
coeffs_highfreq_train = np.concatenate([coeffs_highfreq_train1, coeffs_highfreq_train2, coeffs_highfreq_train3], axis = 0)
# coeffs_lowfreq_validation, coeffs_highfreq_validation = wavelet_transform(X_validation)
# print('X_validation', np.shape(X_validation))
coeffs_lowfreq_test, coeffs_highfreq_test = wavelet_transform(X_test)

[_, low_step_num,low_feature_num] = np.shape(coeffs_lowfreq_train)
[_, high_step_num,high_feature_num] = np.shape(coeffs_highfreq_train)
print('coeffs_lowfreq_train', np.shape(coeffs_lowfreq_train))
print('coeffs_highfreq_train', np.shape(coeffs_highfreq_train))

###############################################GRU模型搭建###############################################################
def adjust_range(x):#调整范围，将每一小段的信号都调整至-0.9~0.9
    max_val = backend.max(x, axis = 1, keepdims=True)#运算Tensor的第一维都是batch，在axis上取均值，为能够广播运算，必须keepdims
    min_val = backend.min(x, axis = 1, keepdims=True)
    y = (x - min_val)/(max_val - min_val + 1e-6)*1.8 - 0.9
    return y

input_lowfreq = layers.Input(shape=[low_step_num, low_feature_num], batch_size = BATCH_SIZE, name='input_lowfreq')
# input_lowfreq_batch = layers.BatchNormalization(axis=0)(input_lowfreq)
input_highfreq = layers.Input(shape=[high_step_num, high_feature_num], batch_size = BATCH_SIZE, name='input_highfreq')
# input_highfreq_batch = layers.BatchNormalization(axis=0)(input_highfreq)
input_label = layers.Input(shape=[SIZE_OUTPUT], batch_size = BATCH_SIZE, name='input_label')

# def biGRU_construct(step_num, feature_size0, feature_size1, feature_size2, feature_size3, compression, name):
# # biGRU model construction
# # input: temporal length, input feature size, hidden feature size, output feature size, downsampling num(int), model name
#     inputs = layers.Input(shape=[step_num, feature_size0], batch_size = BATCH_SIZE, name='input')
#     x = layers.Bidirectional(layers.GRU(feature_size1, return_sequences=True, name='biGRU1'))(inputs)
#     x = layers.Lambda(lambda x:x[:, ::compression, :])(x)
#     x = layers.Bidirectional(layers.GRU(feature_size2, return_sequences=True, name='biGRU2'))(x)
#     outputs = layers.TimeDistributed(layers.Dense(feature_size3, name = 'outputs'))(x)#[batch, LSTM3_INPUT_SIZE, LSTM1_DENSE_SIZE]
#     biGRU_model = models.Model(inputs=inputs, outputs=outputs, name = name)
#     return biGRU_model

def biGRU_construct(step_num, size_input, size_rnn1, size_rnn2, att_size, compression, name):
# biGRU model construction
# input: temporal length, input feature size, hidden feature size, output feature size, downsampling num(int), model name
# 注意力层的输入特征量相等，且与输出特征量相等
    inputs = layers.Input(shape=[step_num, size_input], batch_size = BATCH_SIZE, name='inputs')
    x = layers.Lambda(adjust_range, name = 'adjust_range')(inputs)
    x = layers.Bidirectional(layers.GRU(size_rnn1, return_sequences=True, name='biGRU1'))(x)
    x = layers.Bidirectional(layers.GRU(size_rnn2, return_sequences=True, name='biGRU2'))(x)
    y = layers.Lambda(lambda x:x[:, ::compression, :])(x)
    y = layers.TimeDistributed(layers.Dense(att_size, activation='tanh'))(y)
    inputs_att = layers.TimeDistributed(layers.Dense(att_size, activation='tanh'))(inputs)
    outputs = layers.Attention()([y, inputs_att])
    print('y', y)
    print('inputs_att', inputs_att)
    print('outputs', outputs)
    biGRU_model = models.Model(inputs=inputs, outputs=outputs, name = name)
    return biGRU_model

biGRU_lowfreq = biGRU_construct(np.shape(coeffs_lowfreq_train)[1], np.shape(coeffs_lowfreq_train)[2], 
                                SIZE_RNN1_LOWFREQ, SIZE_RNN2_LOWFREQ, SIZE_LOWFREQ, 1, 'biGRU_lowfreq')
biGRU_highfreq = biGRU_construct(np.shape(coeffs_highfreq_train)[1], np.shape(coeffs_highfreq_train)[2], 
                                 SIZE_RNN1_HIGHFREQ, SIZE_RNN2_HIGHFREQ, SIZE_HIGHFREQ, 4, 'biGRU_highfreq')

feature_lowfreq = biGRU_lowfreq([input_lowfreq])
feature_highfreq = biGRU_highfreq([input_highfreq])
# print('feature_lowfreq', feature_lowfreq)
# print('feature_midfreq', feature_midfreq)
# print('feature_highfreq', feature_highfreq)
feature_freq = layers.Concatenate(axis = 2)([feature_lowfreq, feature_highfreq])
print('feature_freq', feature_freq)

###############################################LSTM模型搭建###############################################################
def biLSTM_construct(step_num, size_input, size_rnn1, att_size, size_rnn2, name):
# biLSTM model construction
# input: temporal length, input feature size, hidden feature size, output feature size, model name
    inputs = layers.Input(shape=[step_num, size_input], batch_size = BATCH_SIZE, name='input')
    x = layers.Bidirectional(layers.LSTM(size_rnn1, return_sequences=True, name='biLSTM1'))(inputs)
    x = layers.TimeDistributed(layers.Dense(att_size, activation='tanh'))(x)
    inputs_att = layers.Dense(att_size, activation='tanh')(inputs)
    x = layers.Attention()([x, inputs_att])
    outputs = layers.Bidirectional(layers.LSTM(size_rnn2, return_sequences=False, name='biLSTM2'))(x)
    biLSTM_model = models.Model(inputs=inputs, outputs=outputs, name = name)
    return biLSTM_model

biLSTM_entire = biLSTM_construct(np.shape(feature_freq)[1], np.shape(feature_freq)[2], 
                                 SIZE_RNN1_ENTIREFREQ, SIZE_ATTENTION_ENTIREFREQ, SIZE_RNN2_ENTIREFREQ, 'biLSTM_entire')
feature = biLSTM_entire([feature_freq])
x = layers.Dense(SIZE_DENSE, activation='tanh')(feature)
x = layers.Dropout(0.5)(x)
output = layers.Dense(SIZE_OUTPUT, name='output', activation='tanh')(x)
predictor = models.Model(inputs=[input_lowfreq, input_highfreq], outputs=output, name='tool_wear_predictor')
# print('output', output)


#############################################模型训练#########################################################################
time_start = time.time()

adam = keras.optimizers.Adam(lr=0.0001)
reduce_lr = keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.2, min_lr=0.00001, mode='auto')

#predictor
predictor.compile(optimizer=adam, loss = 'mae')
print(np.shape(coeffs_lowfreq_train))
print(np.shape(coeffs_highfreq_train))
print(np.shape(Y_train))

#输出的loss表示整体误差，由于decoded和local_decoded用的是同一个模型，误差被整合成了一个
history = predictor.fit([coeffs_lowfreq_train, coeffs_highfreq_train, Y_train], Y_train, validation_split=0.2, callbacks=[reduce_lr],
                        epochs=PREDICTOR_TRAIN_BATCH, batch_size=BATCH_SIZE, shuffle=True, verbose=1)

print('time1 =  ', time.time()-time_start)
```

```python
# 一维离散小波变换
# 对不同传感器信号采用不同的小波级数和频带，降低输入通道数 2020.10.24

import pywt
import numpy as np
import matplotlib.pyplot as plt
from pywt._doc_utils import boundary_mode_subplot
import tensorflow as tf
from tensorflow import keras
import tensorflow.contrib.rnn as rnn
from tensorflow.keras import models, layers, utils, backend
import xlrd
import time
import os

TRAINDATA_LOADPATH = 'tool_wear_data_1/train_data.npy' #训练集数据读取路径
TRAINLABEL_LOADPATH = 'tool_wear_data_1/train_label.npy' #训练集标签读取路径
VALIDATIONDATA_LOADPATH = 'tool_wear_data_1/validation_data.npy' #验证集数据读取路径
VALIDATIONLABEL_LOADPATH =  'tool_wear_data_1/validation_label.npy' #验证集标签读取路径
TESTDATA_LOADPATH = 'tool_wear_data_1/test_data.npy' #验证集数据读取路径
TESTLABEL_LOADPATH =  'tool_wear_data_1/test_label.npy' #验证集标签读取路径
SUMMARY_PATH = './logs'     #记录路径

os.environ["CUDA_VISIBLE_DEVICES"] = "-1"# 这一行注释掉就是使用gpu


#输入数据
INPUT_SIZE = [4096, 7]  #[time_steps, input_vector]
INPUT_NUM = INPUT_SIZE[1]
SCALE_NUM_FORCE = 3 #小波分解级数
SCALE_NUM_VIBRATION = 3 #小波分解级数
SCALE_NUM_AE = 3 #小波分解级数

LOWFREQ_INPUT_SIZE = [512, 7]
LOWFREQ_INPUT_RESHAPESIZE = [16, 32, LOWFREQ_INPUT_SIZE[1]]
LOWFREQ_LSTM_SIZE1 = 25
LOWFREQ_LSTM_SIZE2 = 20
LOWFREQ_DENSE_SIZE = 30

HIGHFREQ_INPUT_SIZE = [512, 18]
HIGHFREQ_INPUT_RESHAPESIZE = [16, 32, HIGHFREQ_INPUT_SIZE[1]]
HIGHFREQ_LSTM_SIZE1 = 25
HIGHFREQ_LSTM_SIZE2 = 20
HIGHFREQ_DENSE_SIZE = 30

TOP_INPUT_SIZE = [LOWFREQ_INPUT_RESHAPESIZE[0], LOWFREQ_DENSE_SIZE + HIGHFREQ_DENSE_SIZE]
TOP_LSTM_SIZE1 = 30
TOP_LSTM_SIZE2 = 25
TOP_DENSE_SIZE = 35

PREDICT_DENSE_SIZE = 250
OUTPUT_SIZE = 3

BATCH_SIZE = 64
PREDICTOR_TRAIN_BATCH = 6

sample_index = 3000

##############################数据集读取########################################
X_train=np.load(TRAINDATA_LOADPATH)
Y_train=np.load(TRAINLABEL_LOADPATH)
# X_validation=np.load(VALIDATIONDATA_LOADPATH)
# Y_validation=np.load(VALIDATIONLABEL_LOADPATH)
X_test=np.load(TESTDATA_LOADPATH)
Y_test=np.load(TESTLABEL_LOADPATH)

#validation_split将样本集按先后比例分为训练集合样本集，样本数量应为两者和的整数倍
X_train = X_train[:(np.shape(X_train)[0]-np.shape(X_train)[0]%np.int(BATCH_SIZE*1.25)), :,:INPUT_NUM]
X_train = X_train[:28160]
Y_train = Y_train[:(np.shape(Y_train)[0]-np.shape(Y_train)[0]%np.int(BATCH_SIZE*1.25)), :OUTPUT_SIZE]
Y_train = Y_train[:28160]

# X_validation = X_validation[:(np.shape(X_validation)[0]-np.shape(X_validation)[0]%BATCH_SIZE), :,:INPUT_NUM]
# Y_validation = Y_validation[:(np.shape(Y_validation)[0]-np.shape(Y_validation)[0]%BATCH_SIZE), :SIZE_OUTPUT]
# print('X_validation', np.shape(X_validation))
# print('Y_validation', np.shape(Y_validation))

#######################################################################################小波分析###############################################################
def percentile_compute(signal, percent): #通过百分数滤除小尺度部分的噪声，输入小波信号整体
    for i in range(np.shape(signal)[2]):
        signal1 = signal[:,:,i]
        percentile = np.percentile(np.abs(signal1), percent)
        signal1[np.abs(signal1) < percentile] = 0
        signal[:,:,i] = signal1
    return signal
def mean_filter(data, ratio):#均值滤波降采样，输入数据和整数压缩比
    index1 = np.int32(np.arange(np.shape(data)[1]* ratio)/ratio)
    if ratio == 0.5:
        return np.mean([np.abs(data[:,index1]), np.abs(data[:,index1+1])], axis = 0)
    if ratio == 0.25:
        return np.mean([np.abs(data[:,index1]), np.abs(data[:,index1+1]), np.abs(data[:,index1+2]), np.abs(data[:,index1+3])], axis = 0)
    else:
        print('error')
def max_filter(data, ratio):#绝对值最大滤波降采样，输入数据和整数压缩比，输出绝对值，如果能输出原值更好
    index1 = np.int32(np.arange(np.shape(data)[1]* ratio)/ratio)
    if ratio == 0.5:
        return np.max([np.abs(data[:,index1]), np.abs(data[:,index1+1])], axis = 0)
    if ratio == 0.25:
        return np.max([np.abs(data[:,index1]), np.abs(data[:,index1+1]), np.abs(data[:,index1+2]), np.abs(data[:,index1+3])], axis = 0)
    else:
        print('error')

def wavelet_transform(data, scale_num):
    coeffs = np.arange(scale_num+1).tolist()
    for j in range(np.shape(data)[2]):
        sample = data[:, :, j].reshape([-1, np.shape(data)[1]])
        coeffs2 = pywt.wavedec(sample, 'db6', 'symmetric', level=scale_num)
        if j == 0:
            for i in range(scale_num+1):
                coeffs[i] = np.array(coeffs2[i]).reshape([np.shape(sample)[0], 1, np.shape(coeffs2[i])[1]])
        else:
            for i in range(scale_num+1):
                coeffs[i] = np.concatenate([coeffs[i], np.array(coeffs2[i]).reshape([np.shape(sample)[0], 1, np.shape(coeffs2[i])[1]])], axis = 1)
    if scale_num == 3:
        coeffs[0] = np.swapaxes(coeffs[0], 1, 2)[:, 4:-5] #低频
        coeffs[1] = np.swapaxes(coeffs[1], 1, 2)[:, 4:-5]
        coeffs[2] = np.swapaxes(coeffs[2], 1, 2)[:, 3:-5]
        coeffs[3] = np.swapaxes(coeffs[3], 1, 2)[:, 2:-3] #高频
    if scale_num == 4:
        coeffs[0] = np.swapaxes(coeffs[0], 1, 2)[:, 4:-6]
        coeffs[1] = np.swapaxes(coeffs[1], 1, 2)[:, 4:-6]
        coeffs[2] = np.swapaxes(coeffs[2], 1, 2)[:, 4:-5]
        coeffs[3] = np.swapaxes(coeffs[3], 1, 2)[:, 3:-5]
        coeffs[4] = np.swapaxes(coeffs[4], 1, 2)[:, 2:-3]
        
    for i in range(scale_num+1):
        print(i, np.shape(coeffs[i]))
    
    for i in range(1, scale_num+1):  #系数显示
        x1 = np.linspace(0, np.shape(coeffs[i])[1]-1, np.shape(coeffs[i])[1])
        x2 = np.linspace(0, np.shape(coeffs[i])[1]-1, np.shape(coeffs[scale_num])[1])
        if i == 1:
            label_interp = np.abs([np.interp(x2, x1, coeffs[i][sample_index, :, 0])])
        else:
            label_interp = np.concatenate([label_interp, np.abs([np.interp(x2, x1, coeffs[i][sample_index, :, 0])])], axis=0)
#     print('label_interp', np.shape(label_interp))
    plt.figure(figsize=(30,2))
    plt.contourf(np.abs(label_interp), cmap='gray')
    plt.show()
    
    return coeffs

coeffs_train_force = wavelet_transform(X_train[:, :, 0:3], SCALE_NUM_FORCE)
# coeffs_train_force[3] = percentile_compute(coeffs_train_force[3], 90) #高频去噪
# coeffs_train_force[2] = percentile_compute(coeffs_train_force[2], 90)
# coeffs_train_force[1] = percentile_compute(coeffs_train_force[1], 90)
coeffs_train_vibration = wavelet_transform(X_train[:, :, 3:6], SCALE_NUM_VIBRATION)
# coeffs_train_vibration[3] = percentile_compute(coeffs_train_vibration[3], 90)
# coeffs_train_vibration[2] = percentile_compute(coeffs_train_vibration[2], 90)
# coeffs_train_vibration[1] = percentile_compute(coeffs_train_vibration[1], 90)
coeffs_train_AE = wavelet_transform(X_train[:, :, 6].reshape([np.shape(X_train)[0], np.shape(X_train)[1], 1]), SCALE_NUM_AE)
coeffs_train_AE = coeffs_train_AE[0]

coeffs512_train = np.concatenate([coeffs_train_force[1], coeffs_train_vibration[1]], axis=-1)
coeffs1024_train = np.concatenate([coeffs_train_force[2], coeffs_train_vibration[2]], axis=-1)
coeffs2048_train = np.concatenate([coeffs_train_force[3], coeffs_train_vibration[3]], axis=-1)
coeffs_lowfreq_train = np.concatenate([coeffs_train_force[0], coeffs_train_vibration[0], coeffs_train_AE], axis = 2)#将不同尺度的小波系数降采样为同一长度
coeffs_highfreq_train = np.concatenate([coeffs512_train, max_filter(coeffs1024_train, 0.5), max_filter(coeffs2048_train, 0.25)], axis = 2)#将不同尺度的小波系数降采样为同一长度

print('coeffs_lowfreq_train', np.shape(coeffs_lowfreq_train))
print('coeffs_highfreq_train', np.shape(coeffs_highfreq_train))

##############################################################################################频带模型搭建###############################################################
def adjust_range(x):#调整范围，将每一小段的信号都调整至-0.9~0.9
    max_val = backend.max(x, axis = 1, keepdims=True)#运算Tensor的第一维都是batch，在axis上取均值，为能够广播运算，必须keepdims
    min_val = backend.min(x, axis = 1, keepdims=True)
    y = (x - min_val)/(max_val - min_val + 1e-6)*1.8 - 0.9
    return y

def freqband_unit_construct(input_size, reshape_size, rnn_size1, rnn_size2, dense_size, name):
# the construction of biLSTM model with attention
# input: input_size, reshape_input_size, hidden_feature_size1, hidden_feature_size2, output_feature_size, model_name
    inputs = layers.Input(shape=input_size, batch_size = BATCH_SIZE, name='input')
    x = layers.Lambda(adjust_range, name = 'adjust_range')(inputs)
    x = layers.Reshape(reshape_size)(x)
    x = layers.TimeDistributed(layers.Bidirectional(layers.LSTM(rnn_size1, return_sequences=True, name='biLSTM1')))(x)
    x = layers.TimeDistributed(layers.Bidirectional(layers.LSTM(rnn_size2, return_sequences=False, name='biLSTM2')))(x)
    outputs = layers.TimeDistributed(layers.Dense(dense_size, activation='tanh'))(x)
    biLSTM_model = models.Model(inputs=inputs, outputs=outputs, name = name)
    return biLSTM_model

input_lowfreq = layers.Input(shape=LOWFREQ_INPUT_SIZE, batch_size = BATCH_SIZE, name='input_lowfreq')
# input_lowfreq_batch = layers.BatchNormalization(axis=0)(input_lowfreq)
input_highfreq = layers.Input(shape=HIGHFREQ_INPUT_SIZE, batch_size = BATCH_SIZE, name='input_highfreq')
# input_highfreq_batch = layers.BatchNormalization(axis=0)(input_highfreq)
biLSTM_lowfreq = freqband_unit_construct(LOWFREQ_INPUT_SIZE, LOWFREQ_INPUT_RESHAPESIZE, LOWFREQ_LSTM_SIZE1, LOWFREQ_LSTM_SIZE2, LOWFREQ_DENSE_SIZE, 'lowfreq_model')
biLSTM_highfreq = freqband_unit_construct(HIGHFREQ_INPUT_SIZE, HIGHFREQ_INPUT_RESHAPESIZE, HIGHFREQ_LSTM_SIZE1, HIGHFREQ_LSTM_SIZE2, HIGHFREQ_DENSE_SIZE, 'highfreq_model')
feature_lowfreq = biLSTM_lowfreq(input_lowfreq)
feature_highfreq = biLSTM_highfreq(input_highfreq)
print('feature_lowfreq', feature_lowfreq)
print('feature_highfreq', feature_highfreq)
feature_freq = layers.Concatenate(axis = 2)([feature_lowfreq, feature_highfreq])
print('feature_freq', feature_freq)

###############################################LSTM模型搭建###############################################################
def top_unit_construct(input_size, rnn_size1, rnn_size2, dense_size, name):
# biLSTM model construction
# input: temporal length, input feature size, hidden feature size, output feature size, model name
# 注意力层的输入特征量相等，且与输出特征量相等
    inputs = layers.Input(shape=input_size, batch_size = BATCH_SIZE, name='input')
    inputs_att = layers.TimeDistributed(layers.Dense(dense_size, activation='tanh'))(inputs)
    print('inputs_att', inputs_att)
    x = layers.Bidirectional(layers.LSTM(rnn_size1, return_sequences=True, name='biLSTM1'))(inputs)
    x = layers.Bidirectional(layers.LSTM(rnn_size2, return_sequences=False, name='biLSTM2'))(x)
    y = layers.Dense(dense_size, activation='tanh')(x)
    print('y', y)
    y = layers.Reshape([1, dense_size])(y)
    print('y', y)
    outputs = layers.Attention()([y, inputs_att])
    print('outputs', outputs)
    outputs = layers.Reshape([dense_size])(outputs)
    biLSTM_model = models.Model(inputs=inputs, outputs=outputs, name = name)
    return biLSTM_model

biLSTM_top = top_unit_construct(TOP_INPUT_SIZE, TOP_LSTM_SIZE1, TOP_LSTM_SIZE2, TOP_DENSE_SIZE, 'biLSTM_entire')
feature = biLSTM_top(feature_freq)
x = layers.Dense(PREDICT_DENSE_SIZE, activation='tanh')(feature)
x = layers.Dropout(0.5)(x)
output = layers.Dense(OUTPUT_SIZE, name='output', activation='tanh')(x)
predictor = models.Model(inputs=[input_lowfreq, input_highfreq], outputs=output, name='tool_wear_predictor')
# print('output', output)

#############################################模型训练#########################################################################
time_start = time.time()

adam = keras.optimizers.Adam(lr=0.0001)
reduce_lr = keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.2, min_lr=0.00001, mode='auto')

#predictor
predictor.compile(optimizer=adam, loss = 'mse')

#输出的loss表示整体误差，由于decoded和local_decoded用的是同一个模型，误差被整合成了一个
history = predictor.fit([coeffs_lowfreq_train, coeffs_highfreq_train], Y_train, validation_split=0.2, callbacks=[reduce_lr],
                        epochs=PREDICTOR_TRAIN_BATCH, batch_size=BATCH_SIZE, shuffle=True, verbose=1)

print('time1 =  ', time.time()-time_start)
```
